```python
import pandas as pd

export_df = pd.read_csv("../dataset/2010_2021_HS2_export.csv")
print(export_df.head())
```

```python
import_df = pd.read_csv("../dataset/2010_2021_HS2_import.csv")
print(import_df.head())
```

```python
print("Export Shape:", export_df.shape)
print("Import Shape:", import_df.shape)
```

```python
print("Export Columns:")
print(export_df.columns)

print("\nImport Columns:")
print(import_df.columns)
```

```python
export_df["Trade_Type"] = "Export"
import_df["Trade_Type"] = "Import"
```

```python
trade_df = pd.concat([export_df, import_df], ignore_index=True)

print("Combined Trade Data Created")
print(trade_df.head())
```

```python
print(trade_df.shape)
print(trade_df.info())
```

```python
trade_df.columns = trade_df.columns.str.lower()
print(trade_df.columns)
```

```python
print(trade_df.isnull().sum())
```

```python
trade_df = trade_df.dropna()
print("After removing nulls:", trade_df.shape)
```

```python
trade_df = trade_df.drop_duplicates()
print("After removing duplicates:", trade_df.shape)
```

```python
trade_df["year"] = trade_df["year"].astype(int)
trade_df["value"] = pd.to_numeric(trade_df["value"])
```

```python
trade_df["country"] = trade_df["country"].str.strip()
trade_df["commodity"] = trade_df["commodity"].str.strip()
trade_df["trade_type"] = trade_df["trade_type"].str.strip()
```

```python
trade_df.to_csv("../dataset/india_trade_cleaned.csv", index=False)

print("✅ Cleaned dataset saved successfully!")
```
